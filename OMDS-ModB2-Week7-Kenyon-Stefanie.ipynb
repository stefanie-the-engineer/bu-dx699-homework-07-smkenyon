Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
# Horizontal Bar Chart (Figure 4.7 on pg. 110)

# Data
categories = ['Engine power is less than expected', 'Tires make excessive noise when driving', 
    'Engine makes abnormal/excessive noise', 'Seat material concerns', 'Excessive wind noise',
    'Hesitation or delay when shifting', 'Bluetooth system has poor sound quality',
    'Steering system/wheel has too much play', 'Bluetooth system is difficult to use',
    'Front seat audio/entertainment/navigation controls']
values = [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]

# Reverse the list data order using slicing [::-1]
categories_reversed = categories[::-1]
values_reversed = values[::-1]

fig, ax = plt.subplots(figsize=(8, 4))

# Create horizontal bar chart
bars = ax.barh(categories_reversed, values_reversed, color='#636363')

# Add a left-aligned graph title
fig.suptitle('Top 10 design concerns', fontsize=14, x=0.02, ha='left')

# Display values inside the bars
ax.bar_label(bars, label_type='edge', padding=-25, color='white', fontsize=10)

# Restore and display the x-axis label on top
ax.xaxis.set_label_position('top')
ax.set_xlabel('concerns per 1,000', fontsize=10, labelpad=0, loc='left')

# Set ticks first, then set the labels exactly once
ax.set_yticks(range(len(categories_reversed)))
ax.set_yticklabels(categories_reversed, fontsize=10)

# Keep x-axis numbers, spines, and ticks removed
ax.set_xticklabels([])

# Remove all spines (top, bottom, left, right)
for spine in ax.spines.values():
    spine.set_visible(False)

# Remove tick marks for a cleaner "no-spine" look
ax.tick_params(axis='both', which='both', length=0)

plt.tight_layout()
plt.show()
```
